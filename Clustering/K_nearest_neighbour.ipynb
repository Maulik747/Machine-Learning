```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing

data=pd.read_csv("car.data")
le=preprocessing.LabelEncoder()
buying=le.fit_transform(list(data['buying']))#returns a numpy array
maint=le.fit_transform(list(data['maint']))
lug_boot=le.fit_transform(list(data['lug_boot']))
safety=le.fit_transform(list(data['safety']))
persons=le.fit_transform(list(data['persons']))
door=le.fit_transform(list(data['door']))
clss=le.fit_transform(list(data['class']))

predict="class"
X=list(zip(buying,maint,lug_boot,safety,persons,door))
y=list(clss)
X_train, X_test, y_train, y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.1)#transform the data in attribute buying to integer values
model=KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
acc=model.score(X_test,y_test)
names=["unacc","acc","good","vgood"]
predicted=model.predict(X_test)
for x in range(len(predicted)):
    print("Predicted: ",names[predicted[x]],"Data: ",X_test[x],"Actual Data: ",names[y_test[x]])


```
