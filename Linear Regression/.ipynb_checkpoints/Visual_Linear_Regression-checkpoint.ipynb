```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
from sklearn import datasets,linear_model
import pickle
from matplotlib import style


stud_data= pd.read_csv(r"C:\Users\DELL\Documents\ML\student-mat.csv", sep=";")
gg= np.array(stud_data['G1'])
newdata= stud_data[['G1','G2','G3',"studytime","failures","absences"]].copy()
predict="G3"
print(newdata.head())
X= np.array(newdata.drop([predict],1))
y=np.array(newdata[predict])
best=0
X_train, X_test, y_train, y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.1)
for _ in range(30):
    X_train, X_test, y_train, y_test=sklearn.model_selection.train_test_split(X,y,test_size=0.1)
    regr= linear_model.LinearRegression()
    regr.fit(X_train, y_train)
    acc=regr.score(X_test,y_test)
    y_predict=regr.predict(X_test)
    print ("accuracy: ",acc)
    if acc>best:# dump only when the accuracy is better than the previous best
        best=acc
        with open("studentmode.pickle","wb")as f:#create a pickle file
            pickle.dump(regr,f) # dump our model in that pickle file

print("The best accuracy is: ",best)
print("coefficient: \n", regr.coef_)#n dimensional coefficient of the n dimensional hyperplane
print("intercept: \n",regr.intercept_) # the intercept
pickle_in=open("studentmode.pickle","rb")
regr=pickle.load(pickle_in)
for x in range (len(y_predict)):
    print(y_predict[x],X_test[x],y_test[x])
print("The accuracy is: ",acc)
print(regr.coef_)
acc=regr.score(X_test,y_test)
print("The accuracy is: ",acc)
style.use("ggplot")
grade1=[]
for i in range(len(X_test)):
    grade1.append(X_test[i][0])
# print(grade1)
print(y_predict)
plt.scatter(grade1,y_test,color='black')
plt.scatter(grade1, y_predict, color='red')
plt.xlabel("grade1")
plt.ylabel("Final Grade")
plt.show()
pickle_in.close()
```
